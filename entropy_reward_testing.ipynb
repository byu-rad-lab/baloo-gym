```python
import numpy as np
from scipy.stats import entropy
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
num_contacts = [1,2,4,8,16,32,64,128,256,512,1024,2048,4096]

# Entropy of a uniform distribution, to see how it scales with number of contacts
entropies = []
for number in num_contacts:
    probabilities = np.ones(number)/number
    entropies.append(entropy(probabilities))

plt.figure()
plt.plot(num_contacts, entropies)
plt.xscale('log')
```

```python
a = np.random.normal(0,scale=50,size=0)
b = np.random.normal(180,scale=80,size=1)
u = np.random.uniform(0,180, 100)

combo = np.concatenate((a,b))
hist, bins = np.histogram(combo, bins=25)
plt.hist(combo, bins=bins)
probabilities = hist/hist.sum()
print(probabilities)
print(entropy(probabilities))
print(entropy(np.ones(len(u))/len(u)))
```
